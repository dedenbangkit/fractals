```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Function to calculate the Sierpiński triangle
def sierpinski_triangle(iterations, points):
    # Define the vertices of the main triangle
    vertices = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3)/2]])
    
    # Initialize the starting point (arbitrary within the triangle)
    current_point = np.array([0.5, 0.5])

    # Store the points to plot
    points_list = [current_point]

    for _ in range(points):
        # Choose a random vertex
        random_vertex = vertices[np.random.choice(len(vertices))]
        
        # Move halfway towards the chosen vertex
        current_point = (current_point + random_vertex) / 2
        points_list.append(current_point)
    
    return np.array(points_list)
```

```python
# Generate Sierpiński triangle points
iterations = 5000
points = sierpinski_triangle(iterations, 20000)
```

```python
# Plot the Sierpiński triangle
plt.figure(figsize=(8, 8))
plt.scatter(points[:, 0], points[:, 1], s=0.1, color="black")
plt.title("Sierpiński Triangle")
plt.axis("equal")
plt.axis("off")
plt.show()
```
