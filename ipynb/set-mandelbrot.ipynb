```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the size of the image and the complex plane
width, height = 800, 800
xmin, xmax = -2.5, 1.5
ymin, ymax = -2.0, 2.0

# Maximum number of iterations
max_iter = 100
```

```python
# Generate the complex plane
x = np.linspace(xmin, xmax, width)
y = np.linspace(ymin, ymax, height)
X, Y = np.meshgrid(x, y)
C = X + 1j * Y
```

```python
# Initialize the iteration array
Z = np.zeros_like(C)
iterations = np.zeros(C.shape, dtype=int)
```

```python
# Iterate the Mandelbrot function
for i in range(max_iter):
    mask = np.abs(Z) <= 2
    Z[mask] = Z[mask] ** 2 + C[mask]
    iterations[mask] = i
```

```python
# Plot the Mandelbrot set
plt.figure(figsize=(10, 10))
plt.imshow(iterations, extent=[xmin, xmax, ymin, ymax], cmap='hot', interpolation='bilinear')
plt.colorbar(label='Iterations')
plt.title("Mandelbrot Set")
plt.xlabel("Real Axis")
plt.ylabel("Imaginary Axis")
plt.show()
```

# Explanation:

- Resolution: The width and height variables determine the resolution of the image.
- Range: The xmin, xmax, ymin, ymax variables define the visible part of the complex plane.
- Coloring: The number of iterations before escaping is used to color each point, creating the fractal's distinctive patterns.


You can modify parameters like the range, resolution, or color map (cmap) to explore different aspects of the Mandelbrot set.
