```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to draw a hexagon (6-sided shape)
def draw_hexagon():
    theta = np.linspace(0, 2 * np.pi, 7)
    x = np.cos(theta)
    y = np.sin(theta)
    return x, y
```

```python
# Plot hexagons with a recursive pattern
plt.figure(figsize=(8, 8))
for scale in range(1, 6):
    x, y = draw_hexagon()
    plt.plot(x * scale, y * scale, linestyle='--', label=f'Scale {scale}')

# Add rotational symmetry lines (9 spokes)
for angle in np.linspace(0, 2 * np.pi, 9, endpoint=False):
    plt.plot([0, np.cos(angle) * 5], [0, np.sin(angle) * 5], color='red', linestyle='-')

plt.title('Hexagonal and Rotational Symmetry')
plt.axis('equal')
plt.legend()
plt.show()
```
