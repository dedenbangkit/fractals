```python
import matplotlib.pyplot as plt

# Function to generate Fibonacci sequence
def fibonacci_sequence(n):
    """
    Generate the Fibonacci sequence up to the n-th term.
    """
    sequence = [0, 1]  # Start with the first two terms
    for i in range(2, n):
        sequence.append(sequence[i - 1] + sequence[i - 2])
    return sequence

# Generate the Fibonacci sequence
n = 20  # Number of terms
fib_sequence = fibonacci_sequence(n)

# Print the sequence
print(f"First {n} terms of the Fibonacci sequence:")
print(fib_sequence)

# Visualize the sequence
plt.figure(figsize=(10, 5))
plt.plot(fib_sequence, marker='o', linestyle='-', color='blue')
plt.title('Fibonacci Sequence')
plt.xlabel('Index')
plt.ylabel('Fibonacci Number')
plt.grid(True)
plt.show()
```

```python
plt.bar(range(n), fib_sequence, color='orange')
plt.title('Fibonacci Sequence as a Bar Chart')
plt.xlabel('Index')
plt.ylabel('Fibonacci Number')
plt.show()
```

```python
golden_ratios = [fib_sequence[i+1] / fib_sequence[i] for i in range(1, len(fib_sequence) - 1)]
print("Golden Ratios:", golden_ratios)
```
