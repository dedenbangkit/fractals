```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the size of the image and the complex plane
width, height = 800, 800
xmin, xmax = -2.0, 2.0
ymin, ymax = -2.0, 2.0

# Maximum number of iterations
max_iter = 100

# Fixed complex parameter (change this to explore different Julia sets)
c = complex(-0.7, 0.27015)
```

```python
# Generate the complex plane
x = np.linspace(xmin, xmax, width)
y = np.linspace(ymin, ymax, height)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y
```

```python
# Initialize the iteration array
iterations = np.zeros(Z.shape, dtype=int)
```

```python
# Iterate the Julia function
for i in range(max_iter):
    mask = np.abs(Z) <= 2
    Z[mask] = Z[mask] ** 2 + c
    iterations[mask] = i
```

```python
# Plot the Julia set
plt.figure(figsize=(10, 10))
plt.imshow(iterations, extent=[xmin, xmax, ymin, ymax], cmap='twilight', interpolation='bilinear')
plt.colorbar(label='Iterations')
plt.title(f"Julia Set (c = {c})")
plt.xlabel("Real Axis")
plt.ylabel("Imaginary Axis")
plt.show()
```

# Explanation:

- Fixed cc: The parameter cc is kept constant, which determines the shape of the Julia set. You can change the value of cc (e.g., c = -0.8 + 0.156j) to see different patterns.
- Complex Plane: The range of the real and imaginary axes (xmin, xmax, ymin, ymax) defines the visible portion of the Julia set.
- Coloring: The number of iterations before escaping is used to assign a color to each point, creating the fractal's intricate design.
